```python
import numpy as np
import matplotlib.pyplot as plt

import scipy.stats as stats
```

```python
# parameters
n1 = 3000
n2 = 4000
mu1 = 1
mu2 = 2

# generate data
data1 = mu1 + np.random.randn(n1)
data2 = mu2 + np.random.randn(n2)

print(data1)
print(data2)
```

```python
# plot data
plt.hist(data1, bins=50, alpha=0.5, label='data1')
plt.hist(data2, bins=50, alpha=0.5, label='data2')
plt.legend()
plt.show()
```

```python
plt.plot(0 + np.random.randn(n1)/15, data1, 'ro', markerfacecolor='w', markersize=14)
plt.plot(1 + np.random.randn(n2)/15, data2, 'bs', markerfacecolor='w', markersize=14)
plt.xlim([-1, 2])
plt.xticks([0, 1], labels=['Group 1', 'Group 2'])
plt.show()
```

```python
t, p = stats.ttest_ind(data1, data2)
print('t =', t)
print('p =', p) # p-value must be less than 0.05 to reject the null hypothesis
```
