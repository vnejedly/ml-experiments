```python
import numpy as np
import matplotlib.pyplot as plt

import sympy as sp
import sympy.plotting.plot as symplot

from IPython.display import display, Math
```

```python
# create symbolic variable
x = sp.symbols('x')

fx = 2*x**2 + 3*x - 5
df = sp.diff(fx, x)

print(f"Derivative of {fx} is {df}")
```

```python
display(fx, df)
symplot(fx, df, (x, -4, 4), title="function")
```

```python
relu = sp.Max(0, x)
sigmoid = 1/(1+sp.exp(-x))

drelu = sp.diff(relu, x)
dsigmoid = sp.diff(sigmoid, x)

display(relu, sigmoid)
symplot(relu, sigmoid, xlim=[-4, 3], ylim=[-1, 1.2], title="activation functions")

display(drelu, dsigmoid)
symplot(drelu, dsigmoid, xlim=[-4, 3], ylim=[-1, 1.2], title="derivatives of activation functions")
```

```python
# evaluate sympy expression

# most simple way
print(f"x=2, y={df.subs(x, 2)}\n")

# lambdify sympy expression
fx_lambda = sp.lambdify(x, fx, "numpy")
df_lambda = sp.lambdify(x, df, "numpy")

print(f"Value of fx at x=2: {fx_lambda(2)}")
print(f"Value of df at x=2: {df_lambda(2)}")
```
